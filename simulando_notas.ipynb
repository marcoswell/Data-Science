# Simulação de Notas

## Projeto de Extensão – Introdução à Estatística

Este código faz parte de um projeto de extensão voltado para o ensino de Estatística a estudantes de escolas públicas do Distrito Federal.

Por meio do conjunto de dados simulados (notas de alunos), os alunos serão introduzidos a conceitos fundamentais como:

- Medidas de tendência central (média, mediana e moda)
- Medidas de dispersão (amplitude, variância e desvio padrão)
- Visualização de dados com gráficos (histogramas, colunas, pizza, etc.)

O objetivo é proporcionar uma experiência prática e acessível, utilizando o **Excel** como ferramenta principal para análise e interpretação dos dados.

```python
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
# Defininado as colunas
colunas = ['Aluno','Turma' , 'Nota - Linguagens', 'Nota - Ciências Humanas', 'Nota - Ciências da Natureza', 'Nota - Matemática', 'Nota - Redação']

#tamanho do Data-Frame
x_tamanho = 100
# Criando um Data-Frame apenas com as colunas
df = pd.DataFrame(columns=colunas, index=range(x_tamanho))
```

```python
# Enumerando os alunos
df['Aluno'] = range(1, x_tamanho + 1)

# Atribuindo as turmas aos estudantes
df['Turma'] = np.random.choice(['A', 'B', 'C'], size=x_tamanho)
```

```python
# Atribuiçao das notas com distribuição 'Normal'

df['Nota - Linguagens'] = np.random.normal(6, 1, x_tamanho)
df['Nota - Ciências Humanas'] = np.random.normal(6.5, 1, x_tamanho)
df['Nota - Ciências da Natureza'] = np.random.normal(7, 0.5, x_tamanho)
df['Nota - Matemática'] = np.random.normal(5.5, 1, x_tamanho)
df['Nota - Redação'] = np.random.normal(7, 1, x_tamanho)
```

```python
df.tail()
```

```python
# Gráfico
plt.figure(figsize=(10, 6))
plt.boxplot(df.iloc[:, 2:7],patch_artist=True)
plt.xticks(range(1, 6), ['Linguagens', 'Ciências Humanas', 'Ciência da Natureza', 'Matemática', 'Redação'])
plt.ylabel('Notas')
plt.title('Distribuição das Notas')
plt.grid(True)
plt.show();
```

```python
# Salvando os arquivos em um csv
df.to_csv('notas_alunos.csv', index=False)
```
