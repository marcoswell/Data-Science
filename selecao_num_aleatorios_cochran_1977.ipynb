# Seleção de Números Aleatórios com Python

Este procedimento utiliza os números aleatórios do **livro de Cochran (1977)** para gerar valores de forma sistemática.

## Critério de Seleção

- A seleção dos números será feita com base em uma **data específica**.
- **Linha inicial**: corresponde ao **dia** da data escolhida.
- **Coluna inicial**: corresponde ao **mês** da data escolhida.
- A partir dessa posição (linha e coluna), os números aleatórios serão extraídos conforme necessário.

## Exemplo de Aplicação

Suponha a data **15/08**:
- Linha selecionada: **15**
- Coluna selecionada: **08**
- O número aleatório inicial será o valor encontrado na interseção da linha 15 com a coluna 8

## Função

```python
import datetime

# Função para a seleção dos números aleatórios 
def selecionar_numeros_aleatorios(n, m, data_amostragem):
    """
    Seleciona números aleatórios de uma tabela de dígitos com base na data atual.

    Args:
        n (int): O número de dígitos para cada número aleatório (tamanho do agrupamento).
        m (int): A quantidade de números aleatórios a serem selecionados.

    Returns:
        list: Uma lista contendo 'm' números aleatórios de 'n' dígitos cada.
    """
    
    # Obter a data de referência
    dia = data_amostragem.day
    mes = data_amostragem.month

    # 2. Limpar e concatenar a tabela em uma única string de dígitos
    # Isso facilita a extração dos números e o "wrap-around" (continuar do início se chegar ao fim)
    digitos_continuos = "".join([linha.replace(" ", "") for linha in tabela_raw])
    
    # 3. Calcular o ponto de partida
    # Garante que o dia e o mês não excedam os limites da tabela
    num_linhas = len(tabela_raw)
    linha_inicial_idx = (dia - 1) % num_linhas 
    
    # Cada linha limpa tem 50 dígitos
    # A posição inicial (coluna) é baseada no mês
    coluna_inicial_idx = (mes - 1)
    
    # O índice de partida na string contínua é o número de linhas completas * 50 + a posição da coluna
    start_index = (linha_inicial_idx * 50) + coluna_inicial_idx

    print(f"Data de referência para o início da seleção: {data_amostragem.strftime('%d/%m/%Y')}")
    print(f"Ponto de partida -> Linha: {linha_inicial_idx + 1}, Posição do dígito: {coluna_inicial_idx + 1}")
    print("-" * 30)

    # 4. Extrair os números
    resultados = []
    current_pos = start_index
    for _ in range(m):
        # Lógica para "dar a volta" na tabela se chegar ao fim
        if current_pos + n > len(digitos_continuos):
            # Pega o que resta da string e completa com o início
            resto = len(digitos_continuos) - current_pos
            numero = digitos_continuos[current_pos:] + digitos_continuos[:n - resto]
            current_pos = n - resto
        else:
            numero = digitos_continuos[current_pos : current_pos + n]
            current_pos += n
        
        resultados.append(numero)
        
    return resultados
```

## Parâmetros

```python
# Parâmetros

# Data de critério para o início da seleção 
data_amostragem = datetime.date(2025, 12, 4)

# n: Quantos dígitos em cada número?
tamanho_agrupamento = 3   # Ex: População de tam. 500 logo teremos amostras de três digítos de 001-500


# m: Quantos números você quer selecionar? - Selecionando mais pois serão excluidos valores repetidos
quantidade_selecionada = 1000 
```

```python

# Coletando os números  
numeros_selecionados = selecionar_numeros_aleatorios(tamanho_agrupamento, quantidade_selecionada, data_amostragem)
    
print(f"Os {quantidade_selecionada} números aleatórios de {tamanho_agrupamento} dígitos são:")
print(numeros_selecionados)
```

## Filtro não podem ter números repetidos ou números maiores que o tamanho amostral 

```python
# excluindo os números repetidos 
print(f"Tamanho original: {len(numeros_selecionados)}")

numeros_selecionados = list(dict.fromkeys(numeros_selecionados)) # mantem a ordem 

print(f"Tamanho após a remoção de duplicados: {len(numeros_selecionados)}")
```

```python
# Excluindo valores maiores que a população 

# tamanho da população 
tam_pop = 473


import numpy as np

numeros_selecionados = np.array(numeros_selecionados, dtype = float)
# Substituindo todos os possíveis 00 por 99
numeros_selecionados[numeros_selecionados == 00] = 99
# Excluindo todos os valores maiores que tam_pop 
numeros_selecionados[numeros_selecionados > tam_pop] = np.nan
# Excluindo os NaN
numeros_selecionados = numeros_selecionados[~np.isnan(numeros_selecionados)]


print(f"Tamanho final: {len(numeros_selecionados)}")

# Selecionando apenas o tamanho da amostra 
tam_amostra = 200
print(f'Tamanho da amostra desejada: {tam_amostra}')
numeros_selecionados = numeros_selecionados[:tam_amostra]
```
