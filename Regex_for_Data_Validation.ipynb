```python
import pandas as pd

# Dados reais costumam vir bagunçados assim:
dados = pd.Series(['$8.500', '9.587', 'oitenta mil', 'R$ 6.578', '05/10/2000'])

# Limpeza com Regex com três casos
limpo = (dados.str.replace(r'[R\$\s]', '', regex=True)     # Remove símbolos e espaços
              .str.replace(r'\.(?=\d{3})', '', regex=True) # Remove ponto de milhar
              .str.replace(',', '.'))                      # Padroniza decimal

# Conversão e Detecção Automática
# O 'coerce' transforma o que não é número em NaN (Not a Number)
df = pd.DataFrame({'Original': dados, 'Convertido': pd.to_numeric(limpo, errors='coerce')})

# Identifica anomalias (o que virou NaN mas não era vazio)
df['Status'] = df['Convertido'].apply(lambda x: 'Ok' if pd.notnull(x) else 'Revisar')
display(df)
```
